```python
from random import randint, choice

#Funções

#printar gráfico e modificá-lo
def plotar_grafico(tab):
    for i, item in enumerate(tab):
        if tab[i] == 'X':
            print('\033[31m', end='')
        if tab[i] == 'O':
            print('\033[34m', end='')
        print(f'{tab[i]:^5}', end='')
        print('\033[m', end='')
        if (i + 1) % 3 == 0:
            print()
            if (i + 1) != len(tab):
                print('-' * 17)
        else:
            print('|', end='')
            
            
def botar_simbolo(tab, escolha_jogador, simbolo_jogador):
    for i, valor in enumerate(tab):
        if valor == escolha_jogador:
            tab[i] = simbolo_jogador
    return tab[:]


#programa
def estrategia_computador(tab, combinacoes, simbolo_jogador, simbolo_computador, primeira_jogada_jogador, segunda_jogada_jogador, primeira_jogada_computador, cont_jogador):              
    #devolvo a primeira jogada do computador
    canto_central = [1, 3, 5, 7]
    canto_esquina = [0, 2, 6, 8]
    if cont_jogador == 1:
        if primeira_jogada_jogador == 4:
            return choice(canto_esquina)
        if primeira_jogada_jogador in canto_central:
            return 4
            if primeira_jogada_jogador == 1:
                posicoes = [0, 2, 4, 7]
                return choice(posicoes)
            if primeira_jogada_jogador == 3:
                posicoes = [0, 4, 5, 6]
                return choice(posicoes)
            if primeira_jogada_jogador== 5:
                posicoes = [2, 3, 4, 8]
                return choice(posicoes)
            if primeira_jogada_jogador == 7:
                posicoes = [1, 4, 6, 8] 
                return choice(posicoes)
        if primeira_jogada_jogador in canto_esquina:
            return 4
        
    #analiso quais combinações podem dar trio na próxima jogada. Vejo 2 símbolos.
    trio_jogador = []
    trio_computador = []
    for combinacao in combinacoes:
        cont_simbolo_jogador = cont_simbolo_computador = 0
        for pos in combinacao:
            if tab[pos] == simbolo_jogador: 
                cont_simbolo_jogador += 1
            if tab[pos] == simbolo_computador:
                cont_simbolo_computador += 1
        if cont_simbolo_jogador == 2 and cont_simbolo_computador == 0:
            trio_jogador.append(combinacao)
        elif cont_simbolo_computador == 2 and cont_simbolo_jogador == 0:
            trio_computador.append(combinacao)
            
    #devolvo jogada se o computador puder formar um trio
    if trio_computador:
        for pos in trio_computador[0]:
            if tab[pos] != simbolo_computador:
                return pos
            
    #devolvo jogada se o jogador puder formar um trio
    if trio_jogador:
        for pos in trio_jogador[0]:
            if tab[pos] != simbolo_jogador:
                return pos
            
    #2º melhor jogada do computador
    if cont_jogador < 3:
        posicao = segunda_jogada_computador(primeira_jogada_jogador, primeira_jogada_computador, segunda_jogada_jogador, canto_esquina, canto_central)
        if posicao or posicao == 0:
            return posicao

    #devolvo posição de um possível trio
    for combinacao in combinacoes:
        lugares_vazios = []
        cont_simb_computador = cont_lug_vazio = 0
        for pos in combinacao:
            if tab[pos] == simbolo_computador:
                cont_simb_computador += 1
            if tab[pos].isnumeric():
                cont_lug_vazio += 1
                lugares_vazios.append(pos)
        if cont_simb_computador == 1 and cont_lug_vazio == 2:
            return choice(lugares_vazios)
        
    #devolvo jogada aleatória
    posicoes_disponiveis = []
    for item in tab:
        if item.isnumeric():
            posicoes_disponiveis.append(tab.index(item))
    return choice(posicoes_disponiveis)


def segunda_jogada_computador(primeira_jogada_jogador, primeira_jogada_computador, segunda_jogada_jogador, canto_esquina, canto_central):
    #jogador joga no meio
    if primeira_jogada_jogador == 4:
        canto_esquina.remove(primeira_jogada_computador) 
        if primeira_jogada_computador == 0 and segunda_jogada_jogador == 8:
            canto_esquina.remove(segunda_jogada_jogador)
            return choice(canto_esquina)
        if primeira_jogada_computador == 2 and segunda_jogada_jogador == 6:
            canto_esquina.remove(segunda_jogada_jogador)
            return choice(canto_esquina)
        if primeira_jogada_computador == 6 and segunda_jogada_jogador == 2:
            canto_esquina.remove(segunda_jogada_jogador)
            return choice(canto_esquina)
        if primeira_jogada_computador == 8 and segunda_jogada_jogador == 0:
            canto_esquina.remove(segunda_jogada_jogador)
            return choice(canto_esquina)
    #jogador joga na esquina
    if primeira_jogada_jogador in canto_esquina:
        if primeira_jogada_jogador == 0 and segunda_jogada_jogador == 8:
            return choice(canto_central)
        if primeira_jogada_jogador == 2 and segunda_jogada_jogador == 6:
            return choice(canto_central)
        if primeira_jogada_jogador == 6 and segunda_jogada_jogador == 2:
            return choice(canto_central)
        if primeira_jogada_jogador == 8 and segunda_jogada_jogador == 0:
            return choice(canto_central)
    #jogador joga no canto central
    if primeira_jogada_jogador in canto_central:
        if primeira_jogada_jogador == 5 and primeira_jogada_computador in [3, 4] and segunda_jogada_jogador in [0, 6]:
            cantos = [2, 8]
            return choice(cantos)
        elif primeira_jogada_jogador == 7 and primeira_jogada_computador in [1, 4] and segunda_jogada_jogador in [0, 2]:
            cantos = [6, 8]
            return choice(cantos)
        elif primeira_jogada_jogador == 3 and primeira_jogada_computador in [5, 4] and segunda_jogada_jogador in [2, 8]:
            cantos = [0, 6]
            return choice(cantos)
        elif primeira_jogada_jogador == 1 and primeira_jogada_computador in [7, 4] and segunda_jogada_jogador in [6, 8]:
            cantos = [0, 2]
            return choice(cantos)

    
def fim(tab, combinacoes, simbolo_jogador, simbolo_computador):
    cont_simbolos = 0
    for valor in tab:
        if valor == simbolo_jogador or valor == simbolo_computador:
            cont_simbolos += 1
    for combinacao in combinacoes:
        cont_simbolo_jogador = cont_simbolo_computador = 0
        for pos in combinacao:
            if tab[pos] == simbolo_jogador: 
                cont_simbolo_jogador += 1
            if tab[pos] == simbolo_computador:
                cont_simbolo_computador += 1
        if cont_simbolo_jogador == 3 or cont_simbolo_computador == 3:
            break
    if cont_simbolo_jogador == 3:
        plotar_grafico(tab)
        print('JOGADOR VENCE! PARABÉNS!!!')
        return True
    elif cont_simbolo_computador == 3:
        plotar_grafico(tab)
        print('COMPUTADOR VENCE! JOGUE NOVAMENTE!!!')
        return True
    elif cont_simbolos == 9:
        plotar_grafico(tab)
        print('DEU VELHA! JOGUE NOVAMENTE!!!')
        return True
    else:
        return False
    
    
#escolhas:
def escolher_pos_jogador(tab):
    while True:
        escolha = str(input('Escreva um número para botar seu símbolo: ')).strip()
        if escolha in tab:
            return escolha
        else:
            print('\033[31mERRO! Escreva um número que tenha na tabela!\033[m')
            
            
def escolher_simbolo_jogador():
    while True:
        escolha = str(input('Você quer X ou O? ')).strip().upper()
        if escolha != 'X' and escolha != 'O':
            print('\033[31mERRO! Escolha X ou O!\033[m')
        else:
            return escolha
        
        
def continuar():
    while True:
        opcao = str(input('Quer continuar? (s/n) ')).strip().casefold()
        if opcao != 's' and opcao != 'n':
            print('\033[31mERRO! Digite S ou N!\033[m')
        else:
            return opcao
            
        
```

```python
#Programa principal

from time import sleep

while True:
    #o jogador escolhe seu símbolo
    simbolo_jogador = escolher_simbolo_jogador()

    #o computador escolhe seu símbolo
    if simbolo_jogador == 'X':
        simbolo_computador = 'O'
    else:
        simbolo_computador = 'X'
        simbolo_jogador = 'O'

    #tabela vazia
    tab = ['1', '2', '3', '4', '5', '6','7', '8', '9']

    #possíveis combinações para vencer
    coluna1 = [0, 3, 6]
    coluna2 = [1, 4, 7]
    coluna3 = [2, 5, 8]
    linha1 = [0, 1, 2]
    linha2 = [3, 4, 5]
    linha3 = [6, 7, 8]
    diagonal1 = [0, 4, 8]
    diagonal2 = [2, 4, 6]
    combinacoes = [coluna1, coluna2, coluna3, linha1, linha2, linha3, diagonal1, diagonal2]

    while True: 
        #plota como está o gráfico
        plotar_grafico(tab)

        #escolho aonde botar o meu simbolo e o programa bota ele
        escolha_jogador = escolher_pos_jogador(tab)
        tab = botar_simbolo(tab, escolha_jogador, simbolo_jogador)
        
        #armazena a primeira e segunda jogada do jogador
        cont_jogador = 0
        for elemento in tab:
            if elemento == simbolo_jogador:
                cont_jogador += 1
        if cont_jogador == 1:
            primeira_jogada_jogador = tab.index(simbolo_jogador) 
            segunda_jogada_jogador = primeira_jogada_computador = 0
        elif cont_jogador == 2:
            for i, valor in enumerate(tab):
                if valor == simbolo_jogador and i != primeira_jogada_jogador:
                    segunda_jogada_jogador = i
                
        #verifico se o jogador venceu ou deu velha
        if fim(tab, combinacoes, simbolo_jogador, simbolo_computador): 
            break
            
        #estratégia e primeira jogada do computador
        escolha_computador = estrategia_computador(tab, combinacoes, simbolo_jogador, simbolo_computador, primeira_jogada_jogador, segunda_jogada_jogador, primeira_jogada_computador, cont_jogador)
        if cont_jogador == 1:
            primeira_jogada_computador = escolha_computador
        tab[escolha_computador] = simbolo_computador
#         print(f'primeira_jogada_jogador = {primeira_jogada_jogador}; segunda_jogada_jogador = {segunda_jogada_jogador}; primeira_jogada_computador = {primeira_jogada_computador}') 
 
        #verifico se o computador venceu ou deu velha
        if fim(tab, combinacoes, simbolo_jogador, simbolo_computador): 
            break
    resposta = continuar()
    if resposta == 'n':
        break
print('\033[31m=' * 16)
print('FIM'.center(16)) 
print('=' * 16 + '\033[m')
```
